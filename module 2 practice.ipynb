```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(x='state', data=df)
plt.show
```

```python
df.sort_values(by="job_openings", inplace=True)
```

```python
comp_df = df.groupby('Industry', as_index=False).agg({'job_openings': 'mean'})
comp_df
```

```python
plt.figure(figsize=(24, 7))
sns.barplot(x= 'Industry', y='job_openings', data=df)
plt.show
```

```python
sns.lineplot(x='founded', y='growth_percentage', data=df)
plt.show()
```

```python
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.show
```

```python
found_df = df.groupby('company_name', as_index=False).agg({'founded': 'mean'})
```

```python
sns.histplot(x='founded', data=df)
plt.show()
```
