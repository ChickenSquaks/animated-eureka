```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
plt.show
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=df)
house_df = df.groupby('YearBuilt', as_index=False).agg({'Price': 'mean'})
plt.show
```

```python
plt.figure(figsize=(18, 8))
sns.barplot(x='Regionname', y='Price', data=df)
houses_df = df.groupby('Regionname', as_index=False).agg({'Price': 'mean'})
plt.show()
```

```python
sns.histplot(x='Price', data=df)
plt.show()
```

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df, orient='h')
plt.show()
```
